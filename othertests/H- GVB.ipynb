# H- with GVB

GVB works for H2, including proper dissociation. But H- is giving a higher energy using GVB than using HF. I suspect there's a sign problem somewhere, but it's hard to see how such a problem wouldn't also affect H2.

```python
%matplotlib inline
from pyquante2.scf.mcscf import gvb
from pyquante2.geo.samples import h
from pyquante2.geo.molecule import molecule
from pyquante2.basis.basisset import basisset
from pyquante2.orbman import orbman
import matplotlib.pyplot as plt
```

```python
Eh = gvb(h)
print Eh
```

```python
h_m = molecule(atomlist = [(1,0,0,0)],charge=-1,name="H-")
E,orbs = gvb(h_m,maxiter=10,basisname='6-31g',npair=0,verbose=True,
             return_orbs=True)
print "HF energy: ", E
o_hf = orbman(orbs,basisset(h_m,'6-31g'),h_m)
E,orbs = gvb(h_m,maxiter=10,basisname='6-31g',npair=1,verbose=True,
             return_orbs=True,input_orbs=orbs)
print "GVB energy: ", E
o_gvb = orbman(orbs,basisset(h_m,'6-31g'),h_m)

for i in [0,1]:
    print("Orbital %d after HF" % i)
    o_hf[i]
    print("  GVB")
    o_gvb[i]
```
